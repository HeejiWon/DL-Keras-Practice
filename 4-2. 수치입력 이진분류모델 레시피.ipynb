__이진분류를 위한 데이터셋을 생성하고 가장 간단한 퍼셉트론 신경망 모델부터 깊은 다층퍼셉트론 신경망 모델까지 구성 및 학습시키기__  
(단, 랜덤 데이터이므로 train set에 과대적합하는 것을 목표로 함)

### 1. 데이터셋 준비

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# 데이터셋 생성

x_train = np.random.random((1000, 12))
y_train = np.random.randint(2, size=(1000, 1))  # 0~1 중 랜덤한 정수
x_test = np.random.random((100, 12))
y_test = np.random.randint(2, size=(100, 1))
```

```python
# 데이터셋 확인 (2차원)
 
plot_x = x_train[:, 0]  # x축(x1), y축(x2)
plot_y = x_train[:, 1]
plot_color = y_train.reshape(1000, )

plt.scatter(plot_x, plot_y, c=plot_color)
plt.show()
```

```python
# 데이터셋 확인 (3차원)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plot_x = x_train[:, 0]
plot_y = x_train[:, 1]
plot_z = x_train[:, 2]
plot_color = y_train.reshape(1000, )

ax.scatter(plot_x, plot_y, plot_z, c=plot_color)
plt.show()
```

### 2. 레이어 준비
- sigmoid

### 3. 모델 준비하기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

#### 1) 퍼셉트론 신경망 모델
- Dense 레이어가 하나, 즉 w와 b 하나로 Y = w * X + b를 풀기 위한 모델 (+sigmoid 함수)

```python
# 2. 모델 구성하기

model = Sequential()
model.add(Dense(1, input_dim=12, activation='sigmoid'))
```

```python
# 3. 모델 학습과정 설정하기

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 4. 모델 학습시키기

hist = model.fit(x_train, y_train, epochs=1000, batch_size=64)
```

```python
# 5. 학습과정 살펴보기

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()  # x축이 같지만 y축이 다른 그래프

loss_ax.set_ylim([0.0, 1.0])
acc_ax.set_ylim([0.0, 1.0])

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

```python
# 6. 모델 평가하기

loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print('loss_and_metrics : ' + str(loss_and_metrics))
```

- activation을 Dense층에 넣지 않고 model.add(Activation('sigmoid'))로 넣을 수 있음

#### 2) 다층퍼셉트론 신경망 모델

```python
# 2. 모델 구성하기

model = Sequential()
model.add(Dense(64, input_dim=12, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 3. 모델 학습과정 설정하기

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 4. 모델 학습시키기

hist = model.fit(x_train, y_train, epochs=1000, batch_size=64)
```

```python
# 5. 학습과정 살펴보기

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.set_ylim([0.0, 1.0])
acc_ax.set_ylim([0.0, 1.0])

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['accuracy'], 'b', label='train acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

```python
# 6. 모델 평가하기

loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print('loss_and_metrics : ' + str(loss_and_metrics))
```

#### 3) 깊은 다층트론 신경망 모델

```python
# 2. 모델 구성하기

model = Sequential()
model.add(Dense(64, input_dim=12, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 3. 모델 학습과정 설정하기

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 4. 모델 학습시키기

hist = model.fit(x_train, y_train, epochs=1000, batch_size=64)
```

```python
# 5. 학습과정 살펴보기

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.set_ylim([0.0, 1.0])
acc_ax.set_ylim([0.0, 1.0])

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

```python
# 6. 모델 평가하기

loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print('loss_and_metrics : ' + str(loss_and_metrics))
```
