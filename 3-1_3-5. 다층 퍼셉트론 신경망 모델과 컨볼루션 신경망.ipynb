### 1. 다층 퍼셉트론 신경망 모델 만들어보기

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
np.random.seed(5)
```

```python
dataset = pd.read_csv("data/diabetes.csv", delimiter=",")
dataset = dataset.values
```

```python
dataset
```

```python
dataset.shape
```

```python
X_train = dataset[:700, 0:8]
y_train = dataset[:700, 8]
X_test = dataset[700:, 0:8]
y_test = dataset[700:, 8]
```

```python
# 모델 구성하기

model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
SVG(model_to_dot(model, show_shapes=True, dpi=70).create(prog='dot', format='svg'))
```

```python
# 모델 학습과정 설정하기

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 모델 학습시키기

model.fit(X_train, y_train, epochs = 1500, batch_size=64)
```

```python
# 모델 평가하기

scores = model.evaluate(X_test, y_test)
print('%s : %.2f%%' %(model.metrics_names[1], scores[1]*100))
```

#### 1) 컨볼루션 신경망 모델 만들어보기
- 손으로 그린 삼각형, 사각형, 원 이미지로 다중 클래스 분류해보기

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.python.keras.utils import np_utils
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
np.random.seed(3)
```

##### 데이터 생성하기

- ImageDataGenerator 클래스를 통해 객체를 생성할 때 파라미터를 전달해주는 것을 통해 데이터의 전처리를 쉽게 할 수 있고, flow_from_directory 메소드를 활용하면 폴더 형태로된 데이터 구조를 바로 가져와서 사용할 수 있음. 데이터 부풀리기(data augmentation)도 가능

```python
train_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(  # 
    'data/handwriting_shape/train',
    target_size = (24, 24),
    batch_size = 3,
    class_mode = 'categorical'  # 다중 클래스 문제이므로
)
```

> - 첫번째 인자 : 이미지의 경로 지정
- target_size : 패치 이미지 크기를 지정 (이 크기로 자동 조절됨)
- batch_size : 배치 크기
- class_mode : 분류 방식<br>
	○ categorical : 2D one-hot 부호화된 라벨이 반환<br>
	○ binary : 1D 이진 라벨이 반환<br>
	○ sparse : 1D 정수 라벨이 반환<br>
    ○ None : 라벨 반환 X

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_generator = test_datagen.flow_from_directory(
    'data/handwriting_shape/test',
    target_size = (24, 24),
    batch_size = 3,
    class_mode = 'categorical'  # 다중 클래스 문제이므로
)
```

##### 모델 구성하기

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu', input_shape=(24, 24, 3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

##### 모델 가시화

```python
SVG(model_to_dot(model, show_shapes=True, dpi=70).create(prog='dot', format='svg'))
```

##### 모델 학습과정 설정하기

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

##### 학습 시키기
- 제너레이터로 생성된 배치를 학습시키는 경우에는 fit_generator() 함수 사용

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=15, # 한 epoch에 사용한 스텝 수 (여기서는 45개의 샘플에 대하여 배치 사이즈가 3이므로)
    epochs=50,
    validation_data=test_generator, # 여기서는 valid set이 따로 없어서 test data 사용
    validation_steps=5  # 한 epoch 종료시마다 사용되는 검증 스텝 수 (여기서 15개의 샘플, 배치 사이즈가 3임)
)
```

##### 모델 평가하기

```python
print('-- Evaluate --')
scores = model.evaluate_generator(test_generator, steps=5) 
print("%s : %.2f%%" %(model.metrics_names[1], scores[1]*100))
```

##### 모델 사용하기

```python
print('-- Predict --')
output = model.predict_generator(test_generator, steps=5)
np.set_printoptions(formatter={'float':lambda x: "{0:0.3f}".format(x)}) # 출력 옵션 변경
                       # {0:0.3f}의 의미 : 0번째 index 값을 소수점 셋째자리까지
print(test_generator.class_indices) # class_indices : 해당열의 클래스 명
print(output)
```

#### 2) 컨볼루션 신경망 모델을 위해 데이터 부풀리기

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
np.random.seed(5)
```

```python
# 데이터 증폭하기

data_aug_gen = ImageDataGenerator(rescale=1./255, 
                                  rotation_range=10,
                                  width_shift_range=0.2,
                                  height_shift_range=0.2,
                                  shear_range=0.7,
                                  zoom_range=[0.9, 2.2],
                                  horizontal_flip=True,
                                  vertical_flip=True,
                                  fill_mode='nearest')
```

> - rotation_range=90 : 0도 ~ 90도 사이에 임의의 각도로 회전시킨
- width_shift_range=0.1 : 전체 넓이의 10%이내만큼 수평방향 이동 (ex. 전체 넓이가 100이면, 10픽셀 내외로 좌우 이동)
- height_shift_range=0.1 : 전체 넓이의 10%이내만큼 수직방향 이동 (ex. 전체 넓이가 100이면, 10픽셀 내외로 상하 이동)
- shear_range=0.5 : 0.5 라디안 내외로 기울임 (회전이 아니라, 특정 각도로 이미지를 늘림)	
- room_range = 0.3 : [1-0.3, 1+0.3] 사이의 범위로 확대/축소 ( x축만 늘리고 줄이는 것 같음 )	
- horizonral_flip=True : 수평방향 뒤집기
- vertical_flip=True : 수직방향 뒤집기
- fill_mode='nearest' : 값이 없는 부분에 대해서는 어떻게 처리할 것인지
	○ nearest : 가장 근처의 값으로 채워넣음
    ○ reflect(반전), wrap(다른 부분 값으로), constant(상수로; 의미없는 값)!

```python
# 이미지가 어떻게 부풀려지는지 하나의 이미지로 확인해보기

img = load_img('data/handwriting_shape/train/triangle/triangle001.png')
x = img_to_array(img)
x.shape        # 마지막 숫자 3은 채널 수
```

```python
x = x.reshape((1,) + x.shape) # (1,) + x.shape : (1, 24, 24, 3)
```

```python
i = 0

# 이 for문은 무한으로 반복되므로 원하는 반복 횟수를 지정하여 빠져나오도록 해야
for batch in train_datagen.flow(x, batch_size=1, save_to_dir='data/handwriting_shape/preview',
                               save_prefix='tri', save_format='png'):
    i += 1
    if i > 30:
        break
```

##### 구현해보기
- 데이터를 부풀리지 않는 경우, steps_per_epoch가 자동으로 설정될 수 있지만, <br>
데이터를 부풀리는 경우, 무한의 샘플이 제공되므로 steps_per_epoch를 지정해야! (여기서는 1500으로 설정)

```python
train_generator = data_aug_gen.flow_from_directory(  # 
    'data/handwriting_shape/train',
    target_size = (24, 24),
    batch_size = 3,
    class_mode = 'categorical'  # 다중 클래스 문제이므로
)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu', input_shape=(24, 24, 3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=15 * 100,   # 1500개 지정 (데이터마다 100개씩 증폭하는 효과)
    epochs=200,
    validation_data=test_generator, # 여기서는 valid set이 따로 없어서 test data 사용
    validation_steps=5  
)
```

```python
# 모델 평가하기

print('-- Evaluate --')
scores = model.evaluate_generator(test_generator, steps=5) 
print("%s : %.2f%%" %(model.metrics_names[1], scores[1]*100))

# 모델 사용하기

print('-- Predict --')
output = model.predict_generator(test_generator, steps=5)
np.set_printoptions(formatter={'float':lambda x: "{0:0.3f}".format(x)}) # 출력 옵션 변경
                       # {0:0.3f}의 의미 : 0번째 index 값을 소수점 셋째자리까지
print(test_generator.class_indices) # class_indices : 해당열의 클래스 명
print(output)
```
