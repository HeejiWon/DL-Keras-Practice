### 1. 데이터셋 준비

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# 데이터셋 생성

x_train = np.random.random((1000, 12))
y_train = np.random.randint(10, size=(1000, 1))
x_test = np.random.random((100, 12))
y_test = np.random.randint(10, size=(100, 1))
```

```python
# 데이터셋 확인 (2차원)

plot_x = x_train[:, 0]
plot_y = x_train[:, 1]
plot_color = y_train.reshape(1000, )

plt.scatter(plot_x, plot_y, c=plot_color)
plt.show()
```

```python
# 데이터셋 확인 (3차원)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plot_x = x_train[:, 0]
plot_y = x_train[:, 1]
plot_z = x_train[:, 2]
plot_color = y_train.reshape(1000, )

ax.scatter(plot_x, plot_y, plot_z, c=plot_color)
plt.show()
```

### 2. 데이터셋 전처리
- one-hot encoding

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

### 3. 레이어 준비
- softmax

### 4. 모델 준비

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 1) 퍼셉트론 신경망 모델
: Dense 레이어 1개

```python
model = Sequential()
model.add(Dense(10, input_dim=12, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
hist = model.fit(x_train, y_train, epochs=1000, batch_size=64)
```

```python
fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()

loss_ax.set_ylim([0.0, 3.0])
acc_ax.set_ylim([0.0, 1.0])

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel("accuracy")

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

```python
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print('loss_and_metrics : ',  str(loss_and_metrics))
```

#### 2) 다층퍼셉트론 신경망 모델
: Dense 레이어 2개

```python
model = Sequential()
model.add(Dense(64, input_dim=12, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
hist = model.fit(x_train, y_train, epochs=1000, batch_size=64)
```

```python
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print('loss_and_metrics : ',  str(loss_and_metrics))
```

#### 3) 깊은 다층퍼셉트론 신경망
: Dense 레이어 3개

```python
model = Sequential()
model.add(Dense(64, input_dim=12, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
hist = model.fit(x_train, y_train, epochs=1000, batch_size=64)
```

```python
fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()

loss_ax.set_ylim([0.0, 3.0])
acc_ax.set_ylim([0.0, 1.0])

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel("accuracy")

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

```python
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print('loss_and_metrics : ',  str(loss_and_metrics))
```
